```python
import random
from datetime import datetime, timedelta
import pandas as pd

random.seed(42)

N_TRIPS = 1000
START_DATE = datetime(2026, 1, 1)
END_DATE = datetime(2026, 12, 31)

trip_types = ["beach", "hiking", "business", "camping", "ski", "city"]
accommodation_types = ["hotel", "hostel", "airbnb", "resort", "tent"]
group_types = ["solo", "couple", "friends", "family"]
budget_ranges = ["budget", "mid", "luxury"]
experience_levels = ["first-time", "occasional", "frequent"]

destinations = [
    ("Bali", "Indonesia", "tropical"),
    ("Phuket", "Thailand", "tropical"),
    ("Goa", "India", "tropical"),
    ("Zurich", "Switzerland", "temperate"),
    ("Manali", "India", "mountain"),
    ("Shimla", "India", "mountain"),
    ("Tokyo", "Japan", "temperate"),
    ("London", "UK", "temperate"),
    ("Dubai", "UAE", "arid"),
    ("Singapore", "Singapore", "tropical"),
    ("Bangkok", "Thailand", "tropical"),
    ("Paris", "France", "temperate"),
]

departure_countries = ["India", "USA", "UK", "Germany", "Australia", "Canada", "UAE"]

activity_map = {
    "beach": ["swimming", "sunbathing", "snorkeling", "sightseeing", "nightlife"],
    "hiking": ["trekking", "trail-walk", "campfire", "nature-tour"],
    "business": ["meetings", "conference", "formal-dinners", "city-tour"],
    "camping": ["camping", "bonfire", "stargazing", "trail-walk"],
    "ski": ["skiing", "snow-walk", "mountain-tour"],
    "city": ["sightseeing", "shopping", "city-tour", "museum-visit"],
}

catalog_rows = [
    ("ITM0001","tshirt","clothes","top",0),("ITM0002","cotton_shirt","clothes","top",0),
    ("ITM0003","shorts","clothes","bottom",0),("ITM0004","jeans","clothes","bottom",0),
    ("ITM0005","trousers","clothes","bottom",0),("ITM0006","swimwear","clothes","beachwear",0),
    ("ITM0007","light_jacket","clothes","outerwear",0),("ITM0008","rain_jacket","clothes","outerwear",0),
    ("ITM0009","thermal_top","clothes","winterwear",0),("ITM0010","thermal_bottom","clothes","winterwear",0),
    ("ITM0011","hoodie","clothes","winterwear",0),("ITM0012","sweater","clothes","winterwear",0),
    ("ITM0013","socks","clothes","innerwear",0),("ITM0014","innerwear","clothes","innerwear",0),
    ("ITM0015","sleepwear","clothes","nightwear",0),("ITM0016","formal_shirt","clothes","formal",0),
    ("ITM0017","formal_trousers","clothes","formal",0),("ITM0018","blazer","clothes","formal",0),
    ("ITM0019","sandals","shoes","beach",0),("ITM0020","flip_flops","shoes","beach",0),
    ("ITM0021","sneakers","shoes","casual",0),("ITM0022","hiking_boots","shoes","outdoor",0),
    ("ITM0023","snow_boots","shoes","winter",0),("ITM0024","formal_shoes","shoes","formal",0),
    ("ITM0025","sunglasses","accessories","sun",0),("ITM0026","sun_hat","accessories","sun",0),
    ("ITM0027","cap","accessories","casual",0),("ITM0028","scarf","accessories","winter",0),
    ("ITM0029","gloves","accessories","winter",0),("ITM0030","belt","accessories","formal",0),
    ("ITM0031","watch","accessories","casual",0),("ITM0032","toothbrush","toiletries","oral-care",0),
    ("ITM0033","toothpaste","toiletries","oral-care",0),("ITM0034","shampoo","toiletries","hair-care",0),
    ("ITM0035","soap","toiletries","body-care",0),("ITM0036","sunscreen","toiletries","skin-care",0),
    ("ITM0037","moisturizer","toiletries","skin-care",0),("ITM0038","lip_balm","toiletries","skin-care",0),
    ("ITM0039","deodorant","toiletries","body-care",0),("ITM0040","hand_sanitizer","toiletries","hygiene",0),
    ("ITM0041","towel","toiletries","bath",0),("ITM0042","backpack","gear","bag",0),
    ("ITM0043","daypack","gear","bag",0),("ITM0044","power_bank","gear","electronics",0),
    ("ITM0045","phone_charger","gear","electronics",0),("ITM0046","universal_adapter","gear","electronics",0),
    ("ITM0047","water_bottle","gear","utility",0),("ITM0048","umbrella","gear","weather",0),
    ("ITM0049","rain_cover","gear","weather",0),("ITM0050","trekking_pole","gear","outdoor",0),
    ("ITM0051","headlamp","gear","outdoor",0),("ITM0052","sleeping_bag","gear","camping",0),
    ("ITM0053","painkiller","medicines","basic",1),("ITM0054","bandage","medicines","basic",1),
    ("ITM0055","antacid","medicines","basic",1),("ITM0056","ors","medicines","basic",1),
    ("ITM0057","motion_sickness_tablet","medicines","basic",1),("ITM0058","cough_lozenge","medicines","basic",1),
]
item_catalog = pd.DataFrame(catalog_rows, columns=["item_id","item_name","category","subcategory","is_basic_medicine"])
item_by_name = {r[1]: r[0] for r in catalog_rows}
cat_by_item = {r[0]: r[2] for r in catalog_rows}

def random_date(start, end):
    return start + timedelta(days=random.randint(0, (end-start).days))

def get_season(dt):
    if dt.month in [12,1,2]: return "winter"
    if dt.month in [3,4,5]: return "spring"
    if dt.month in [6,7,8]: return "summer"
    return "autumn"

def temp_rain_humidity(climate, season):
    if climate == "tropical": return random.randint(27,33), round(random.uniform(0.35,0.75),2), "high"
    if climate == "arid": return random.randint(24,40), round(random.uniform(0.01,0.15),2), "low"
    if climate == "mountain":
        t = random.randint(-5,8) if season=="winter" else random.randint(8,22)
        return t, round(random.uniform(0.15,0.55),2), "medium"
    if season == "winter": t = random.randint(-2,10)
    elif season == "summer": t = random.randint(18,30)
    else: t = random.randint(8,20)
    return t, round(random.uniform(0.15,0.55),2), random.choice(["low","medium"])

def pick_items(trip_type, season, climate, group_type, has_kids):
    base = {"tshirt","innerwear","socks","toothbrush","toothpaste","soap","shampoo",
            "deodorant","backpack","phone_charger","power_bank","water_bottle",
            "painkiller","bandage","antacid"}
    if climate == "tropical": base |= {"sunscreen","sunglasses","cap"}
    if season == "winter": base |= {"sweater","hoodie","lip_balm"}
    if trip_type == "beach": base |= {"swimwear","shorts","sandals","flip_flops","sun_hat"}
    elif trip_type == "hiking": base |= {"hiking_boots","rain_jacket","trekking_pole","daypack","headlamp"}
    elif trip_type == "business": base |= {"formal_shirt","formal_trousers","formal_shoes","belt","universal_adapter"}
    elif trip_type == "camping": base |= {"hiking_boots","sleeping_bag","headlamp","rain_jacket"}
    elif trip_type == "ski": base |= {"thermal_top","thermal_bottom","snow_boots","gloves","scarf","hoodie"}
    elif trip_type == "city": base |= {"sneakers","cotton_shirt","jeans","daypack"}
    if has_kids: base |= {"ors"}
    if group_type == "family": base |= {"hand_sanitizer","towel"}
    for _ in range(random.randint(1,4)):
        base.add(random.choice(["umbrella","rain_cover","motion_sickness_tablet","cough_lozenge","moisturizer","watch","sleepwear"]))
    return list(base)

def quantity_priority(item_name, duration):
    if item_name in ["tshirt","innerwear","socks"]: return max(3,min(10,duration)), "high", 1
    if item_name in ["toothbrush","toothpaste","phone_charger","painkiller","bandage"]: return 1, "high", 1
    if item_name in ["swimwear","formal_shirt","formal_trousers","thermal_top","thermal_bottom"]: return max(1,duration//3), "high", 1
    pr = random.choice(["low","medium","high"])
    return random.randint(1,2), pr, 1 if (pr=="high" and random.random()<0.7) else 0

def tip_templates(trip_type, climate, season):
    tips=[]
    if climate=="tropical": tips += ["Carry breathable cotton clothes for humid weather.","Reapply sunscreen every 3-4 hours."]
    if climate=="arid": tips += ["Hydrate frequently and use moisturizer in dry conditions."]
    if season=="winter": tips += ["Use layering to adapt to temperature changes."]
    if trip_type=="beach": tips += ["Keep an extra dry set of clothes after water activities."]
    elif trip_type=="hiking": tips += ["Wear proper ankle-support footwear on trails."]
    elif trip_type=="business": tips += ["Keep one wrinkle-free formal outfit ready."]
    elif trip_type=="camping": tips += ["Keep a flashlight/headlamp easily accessible at night."]
    elif trip_type=="ski": tips += ["Use thermal base layers and waterproof outerwear."]
    elif trip_type=="city": tips += ["Carry a compact daypack for daily city travel."]
    return tips[:3]

trips, labels, tips_rows = [], [], []

for i in range(1, N_TRIPS+1):
    trip_id = f"TRIP{i:06d}"
    city, dest_country, climate = random.choice(destinations)
    dep_country = random.choice(departure_countries)
    trip_type = random.choice(trip_types)
    accom = random.choice(accommodation_types)
    group = random.choice(group_types)
    budget = random.choice(budget_ranges)
    exp = random.choice(experience_levels)

    start = random_date(START_DATE, END_DATE)
    duration = random.randint(2,12)
    end = start + timedelta(days=duration)
    season = get_season(start)

    has_toddlers = 1 if group=="family" and random.random()<0.25 else 0
    has_kids = 1 if group=="family" and random.random()<0.45 else 0
    has_seniors = 1 if group=="family" and random.random()<0.20 else 0
    headcount = 1 if group=="solo" else 2 if group=="couple" else random.randint(3,6 if group=="friends" else 7)

    avg_temp, rain_prob, humidity = temp_rain_humidity(climate, season)
    acts = random.sample(activity_map[trip_type], k=2)
    baggage = random.choice([15,20,25,30])

    trips.append({
        "trip_id":trip_id,"start_date":start.strftime("%Y-%m-%d"),"end_date":end.strftime("%Y-%m-%d"),
        "duration_days":duration,"season":season,"destination_city":city,"destination_country":dest_country,
        "departure_country":dep_country,"climate_zone":climate,"trip_type":trip_type,
        "accommodation_type":accom,"group_type":group,"has_toddlers":has_toddlers,"has_kids":has_kids,
        "has_seniors":has_seniors,"total_headcount":headcount,"budget_range":budget,
        "trip_experience_level":exp,"avg_temp_c":avg_temp,"rain_probability":rain_prob,
        "humidity_level":humidity,"planned_activities":",".join(acts),"baggage_limit_kg":baggage
    })

    for it in pick_items(trip_type, season, climate, group, has_kids):
        if it in item_by_name:
            item_id = item_by_name[it]
            qty, pri, mand = quantity_priority(it, duration)
            labels.append({
                "trip_id":trip_id,"item_id":item_id,"category":cat_by_item[item_id],
                "quantity":qty,"priority":pri,"is_mandatory":mand,
                "reason":f"{trip_type} trip in {climate} climate during {season}"
            })

    for t_idx, tip in enumerate(tip_templates(trip_type, climate, season), start=1):
        tips_rows.append({
            "trip_id":trip_id,
            "tip_id":f"TIP{trip_id[-6:]}_{t_idx}",
            "tip_text":tip,
            "priority":"high" if t_idx==1 else "medium"
        })

trips_df = pd.DataFrame(trips)
labels_df = pd.DataFrame(labels)
tips_df = pd.DataFrame(tips_rows)

trips_df.to_csv("/content/trips.csv", index=False)
item_catalog.to_csv("/content/item_catalog.csv", index=False)
labels_df.to_csv("/content/trip_item_labels.csv", index=False)
tips_df.to_csv("/content/trip_tips_labels.csv", index=False)

print("✅ Files created in /content:")
print("trips.csv:", len(trips_df))
print("item_catalog.csv:", len(item_catalog))
print("trip_item_labels.csv:", len(labels_df))
print("trip_tips_labels.csv:", len(tips_df))
```

```python
# 2) Imports
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder, MultiLabelBinarizer
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.multiclass import OneVsRestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import f1_score
```

```python
# 3) Load data
trips = pd.read_csv('/content/trips.csv')
items = pd.read_csv('/content/item_catalog.csv')
labels = pd.read_csv('/content/trip_item_labels.csv')
tips = pd.read_csv('/content/trip_tips_labels.csv')

print(trips.shape, items.shape, labels.shape, tips.shape)
```

```python
# 4) Build Y (multi-label by item_name)
labels2 = labels.merge(items[['item_id','item_name']], on='item_id', how='left')
trip_to_items = labels2.groupby('trip_id')['item_name'].apply(list).reset_index()

df = trips.merge(trip_to_items, on='trip_id', how='left')
df['item_name'] = df['item_name'].apply(lambda x: x if isinstance(x, list) else [])

mlb = MultiLabelBinarizer()
Y = mlb.fit_transform(df['item_name'])   # target matrix
item_classes = mlb.classes_
```

```python
# 5) Build X features
feature_cols = [
    'duration_days','season','destination_country','departure_country','climate_zone',
    'trip_type','accommodation_type','group_type','has_toddlers','has_kids','has_seniors',
    'total_headcount','budget_range','trip_experience_level','avg_temp_c',
    'rain_probability','humidity_level','baggage_limit_kg'
]

X = df[feature_cols].copy()

cat_cols = [
    'season','destination_country','departure_country','climate_zone','trip_type',
    'accommodation_type','group_type','budget_range','trip_experience_level','humidity_level'
]
num_cols = [c for c in feature_cols if c not in cat_cols]

preprocessor = ColumnTransformer(
    transformers=[
        ('cat', OneHotEncoder(handle_unknown='ignore'), cat_cols),
        ('num', 'passthrough', num_cols)
    ]
)
```

```python
# 6) Train/test split
X_train, X_test, Y_train, Y_test = train_test_split(
    X, Y, test_size=0.2, random_state=42
)

model = Pipeline([
    ('prep', preprocessor),
    ('clf', OneVsRestClassifier(LogisticRegression(max_iter=200)))
])

model.fit(X_train, Y_train)
Y_pred = model.predict(X_test)

print("F1 micro:", f1_score(Y_test, Y_pred, average='micro'))
print("F1 macro:", f1_score(Y_test, Y_pred, average='macro'))
```

```python
item_to_cat = dict(zip(items['item_name'], items['category']))

def build_output_for_trip(trip_row, threshold=0.35):
    probs = model.predict_proba(pd.DataFrame([trip_row]))[0]
    picked = [item_classes[i] for i,p in enumerate(probs) if p >= threshold]

    if len(picked) < 8:
        top_idx = np.argsort(probs)[::-1][:10]
        picked = list(dict.fromkeys([item_classes[i] for i in top_idx]))

    out = {
        "trip_id": trip_row.get("trip_id", "NEW_TRIP"),
        "clothes": [], "shoes": [], "accessories": [],
        "toiletries": [], "gear": [], "medicines": [],
        "personalized_tips": []
    }

    for it in picked:
        cat = item_to_cat.get(it, None)
        if cat == "clothes": out["clothes"].append(it)
        elif cat == "shoes": out["shoes"].append(it)
        elif cat == "accessories": out["accessories"].append(it)
        elif cat == "toiletries": out["toiletries"].append(it)
        elif cat == "gear": out["gear"].append(it)
        elif cat == "medicines": out["medicines"].append(it)

    if trip_row['trip_type'] == 'beach':
        out["personalized_tips"].append("Carry breathable beachwear and quick-dry clothes.")
    if trip_row['climate_zone'] == 'tropical':
        out["personalized_tips"].append("Use sunscreen and hydrate frequently.")
    if trip_row['season'] == 'winter':
        out["personalized_tips"].append("Use layered clothing for temperature variation.")
    if trip_row['trip_type'] == 'business':
        out["personalized_tips"].append("Keep one formal outfit wrinkle-free.")

    basic_meds = {"painkiller","bandage","antacid","ors","motion_sickness_tablet","cough_lozenge"}
    out["medicines"] = [m for m in out["medicines"] if m in basic_meds]

    return out
```

```python
# Manual user input -> prediction

def predict_from_manual_input(user_input, threshold=0.35):
    """
    user_input: dict with same keys as training feature_cols (+ optional trip_id)
    """
    required = set(feature_cols)
    missing = required - set(user_input.keys())
    if missing:
        raise ValueError(f"Missing fields: {missing}")

    row_df = pd.DataFrame([user_input])[feature_cols]
    probs = model.predict_proba(row_df)[0]

    picked = [item_classes[i] for i, p in enumerate(probs) if p >= threshold]
    if len(picked) < 8:  # fallback for minimum useful list
        top_idx = np.argsort(probs)[::-1][:10]
        picked = list(dict.fromkeys([item_classes[i] for i in top_idx]))

    result = {
        "trip_id": user_input.get("trip_id", "MANUAL_TRIP_001"),
        "clothes": [],
        "shoes": [],
        "accessories": [],
        "toiletries": [],
        "gear": [],
        "medicines": [],
        "personalized_tips": []
    }

    for it in picked:
        cat = item_to_cat.get(it)
        if cat == "clothes": result["clothes"].append(it)
        elif cat == "shoes": result["shoes"].append(it)
        elif cat == "accessories": result["accessories"].append(it)
        elif cat == "toiletries": result["toiletries"].append(it)
        elif cat == "gear": result["gear"].append(it)
        elif cat == "medicines": result["medicines"].append(it)

    # rule-based safety/quality boosts
    if user_input["trip_type"] == "beach" and user_input["climate_zone"] == "tropical":
        for must_item, cat in [("sunscreen", "toiletries"), ("sunglasses", "accessories"), ("swimwear", "clothes")]:
            if must_item not in result[cat]:
                result[cat].append(must_item)

    if user_input["trip_type"] == "ski":
        for must_item, cat in [("thermal_top", "clothes"), ("snow_boots", "shoes"), ("gloves", "accessories")]:
            if must_item not in result[cat]:
                result[cat].append(must_item)

    # medicines: keep only basic
    basic_meds = {"painkiller","bandage","antacid","ors","motion_sickness_tablet","cough_lozenge"}
    result["medicines"] = [m for m in result["medicines"] if m in basic_meds]

    # tips
    if user_input["climate_zone"] == "tropical":
        result["personalized_tips"].append("Carry breathable cotton clothes and stay hydrated.")
    if user_input["trip_type"] == "beach":
        result["personalized_tips"].append("Reapply sunscreen every 3-4 hours.")
    if user_input["season"] == "winter":
        result["personalized_tips"].append("Use layered clothing for better warmth control.")
    if user_input["trip_type"] == "business":
        result["personalized_tips"].append("Keep one wrinkle-free formal outfit ready.")

    return result
```

```python
import pandas as pd
import numpy as np

# -----------------------------
# 1) Destination -> climate map
#    (extend this dictionary)
# -----------------------------
DEST_CLIMATE_MAP = {
    ("bali", "indonesia"): "tropical",
    ("phuket", "thailand"): "tropical",
    ("goa", "india"): "tropical",
    ("singapore", "singapore"): "tropical",
    ("bangkok", "thailand"): "tropical",
    ("zurich", "switzerland"): "temperate",
    ("london", "uk"): "temperate",
    ("paris", "france"): "temperate",
    ("tokyo", "japan"): "temperate",
    ("manali", "india"): "mountain",
    ("shimla", "india"): "mountain",
    ("dubai", "uae"): "arid",
}

# fallback by country (if city not found)
COUNTRY_CLIMATE_MAP = {
    "indonesia": "tropical",
    "thailand": "tropical",
    "singapore": "tropical",
    "india": "temperate",   # generic fallback
    "switzerland": "temperate",
    "uk": "temperate",
    "france": "temperate",
    "japan": "temperate",
    "uae": "arid",
}

def derive_season(date_str):
    dt = pd.to_datetime(date_str)
    m = dt.month
    if m in [12, 1, 2]:
        return "winter"
    elif m in [3, 4, 5]:
        return "spring"
    elif m in [6, 7, 8]:
        return "summer"
    else:
        return "autumn"

def derive_climate_zone(destination_city, destination_country):
    city = str(destination_city).strip().lower()
    country = str(destination_country).strip().lower()

    if (city, country) in DEST_CLIMATE_MAP:
        return DEST_CLIMATE_MAP[(city, country)]
    return COUNTRY_CLIMATE_MAP.get(country, "temperate")  # safe default

def build_feature_row_from_user_input(user_input):
    """
    user_input must contain only your listed fields:
    trip_id, start_date, end_date, destination_city, destination_country, departure_country,
    trip_type, accommodation_type, group_type, has_toddlers, has_kids, has_seniors,
    total_headcount, budget_range, trip_experience_level
    """
    start = pd.to_datetime(user_input["start_date"])
    end = pd.to_datetime(user_input["end_date"])

    duration_days = (end - start).days
    if duration_days <= 0:
        raise ValueError("end_date must be after start_date")

    season = derive_season(user_input["start_date"])
    climate_zone = derive_climate_zone(user_input["destination_city"], user_input["destination_country"])

    # If your trained model needs these extra columns, set defaults
    avg_temp_c_default = 27 if climate_zone == "tropical" else 15 if climate_zone == "temperate" else 5 if climate_zone == "mountain" else 32
    rain_probability_default = 0.5 if climate_zone == "tropical" else 0.3 if climate_zone == "temperate" else 0.35 if climate_zone == "mountain" else 0.1
    humidity_level_default = "high" if climate_zone == "tropical" else "medium" if climate_zone in ["temperate", "mountain"] else "low"
    baggage_limit_default = 20

    row = {
        "trip_id": user_input["trip_id"],
        "duration_days": duration_days,
        "season": season,
        "destination_country": user_input["destination_country"],
        "departure_country": user_input["departure_country"],
        "climate_zone": climate_zone,
        "trip_type": user_input["trip_type"],
        "accommodation_type": user_input["accommodation_type"],
        "group_type": user_input["group_type"],
        "has_toddlers": int(user_input["has_toddlers"]),
        "has_kids": int(user_input["has_kids"]),
        "has_seniors": int(user_input["has_seniors"]),
        "total_headcount": int(user_input["total_headcount"]),
        "budget_range": user_input["budget_range"],
        "trip_experience_level": user_input["trip_experience_level"],
        "avg_temp_c": avg_temp_c_default,
        "rain_probability": rain_probability_default,
        "humidity_level": humidity_level_default,
        "baggage_limit_kg": baggage_limit_default
    }
    return row

def predict_from_minimal_input(user_input, threshold=0.35):
    # Build derived feature row
    row = build_feature_row_from_user_input(user_input)

    # Ensure same training columns
    row_df = pd.DataFrame([row])[feature_cols]

    probs = model.predict_proba(row_df)[0]
    picked = [item_classes[i] for i, p in enumerate(probs) if p >= threshold]
    if len(picked) < 8:
        top_idx = np.argsort(probs)[::-1][:10]
        picked = [item_classes[i] for i in top_idx]

    result = {
        "trip_id": row["trip_id"],
        "clothes": [], "shoes": [], "accessories": [],
        "toiletries": [], "gear": [], "medicines": [],
        "personalized_tips": []
    }

    for it in picked:
        cat = item_to_cat.get(it)
        if cat == "clothes": result["clothes"].append(it)
        elif cat == "shoes": result["shoes"].append(it)
        elif cat == "accessories": result["accessories"].append(it)
        elif cat == "toiletries": result["toiletries"].append(it)
        elif cat == "gear": result["gear"].append(it)
        elif cat == "medicines": result["medicines"].append(it)

    # Rule boosts
    if row["trip_type"] == "beach" and row["climate_zone"] == "tropical":
        for it, cat in [("swimwear","clothes"), ("sunglasses","accessories"), ("sunscreen","toiletries"), ("sandals","shoes")]:
            if it not in result[cat]:
                result[cat].append(it)

    # basic medicines only
    basic_meds = {"painkiller","bandage","antacid","ors","motion_sickness_tablet","cough_lozenge"}
    result["medicines"] = [m for m in result["medicines"] if m in basic_meds]

    # tips
    result["personalized_tips"].append(f"Trip duration is {row['duration_days']} days; pack accordingly.")
    if row["climate_zone"] == "tropical":
        result["personalized_tips"].append("Carry light breathable clothes and stay hydrated.")
    if row["trip_type"] == "beach":
        result["personalized_tips"].append("Use sunscreen and sunglasses for strong sun exposure.")

    return result
```
